```python
vectors = {}

with open("/home/sofialyubina/tools/word2vec/bin/vectors_very_big_book.processed", "r") as infile:
    infile.readline()
    
    for line in infile:
        elements = line.strip().split(" ")
        word = elements[0]
        vector = []
        for element in elements[1:]:
            vector.append(float(element))
            
        vectors[word] = np.array(vector)
```

```python
def calc_vector_len(vector):
    return np.sqrt(np.sum(vector * vector))
```

```python
def get_words(text):
    text = text.lower()
    text = text.replace(",", " ")
    text = text.replace(".", " ")
    new_text = text.split(" ")
    
    return new_text

def middle_vector(a,vectors):
    result = np.zeros(300)
    
    words = get_words(a)
    number = 0
    
    for n in range(len(words)):
        word = words[n]
        if word in vectors:
            result  = vectors[word] +  result
            number += 1
    
    if number > 0:
        result /= number
    
    return result

def vectors_distance(first, second, vectors):
    a = middle_vector(first, vectors)
    b = middle_vector(second, vectors)
    
    result = np.sum(a * b) / calc_vector_len(a) / calc_vector_len(b)
    return result
```
